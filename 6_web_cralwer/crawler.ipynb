```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd


def mover():
    url = "https://race.kra.co.kr/raceScore/ScoretableScoreList.do?Act=04&Sub=1&meet=1"
    raw_text = requests.get(url)
    #print(raw_text.text)
    soup = BeautifulSoup(raw_text.text, 'lxml')
    main_table = soup.findAll('table', attrs={'summary':'순서,경주일자,경주 순으로 정보제공'})
    #print(type(main_table))
    for i, td in enumerate(main_table):

        main_table_href = td.findAll('a')
        for j, href in enumerate(main_table_href):
            if(j==1): # for test
                # 정규식 표현
                pattern = "'(\d+)','(\d+)','(\d+)'"
                r = re.compile(pattern)
                print(str(href))
                if(r.search(str(href))):
                    match = r.search(str(href))
                    print(match.group(1))
                    print(match.group(2))
                    print(match.group(3))
                    meet = match.group(1)
                    realRcDate = match.group(2)
                    realRcN = match.group(3)
                    goSuburl(meet ,realRcDate,realRcN)
                        #Make url 
            
def goSuburl(meet,realRcDate,realRcN ):
    data = []
    dataheader = []
    url = "https://race.kra.co.kr/raceScore/ScoretableDetailList.do?meet="+str(meet)+"&realRcDate="+str(realRcDate) + "&realRcNo=" +str(realRcN)
    raw_text = requests.get(url)
    soup = BeautifulSoup(raw_text.text, 'lxml')
    print(type(soup))
    sub_div_table = soup.findAll('div', attrs={'class':'tableType2'})
    for k, table_content in enumerate(sub_div_table):
        if(k==0):
            table_body = table_content.find('table')
            print(type(table_body))
            table_tbody_h = table_body.find('thead')
            hrows = table_tbody_h.find_all('tr')
            for rowh in hrows:
                colsh = rowh.find_all('th')
                colsh = [eleh.text.strip() for eleh in colsh]
                dataheader.append([eleh for eleh in colsh if eleh]) # Get rid of empty values
                
            print(dataheader)
            table_tbody2 = table_body.find('tbody')
            #print(table_tbody2)
            #print(type(table_tbody2))
            rows = table_body.find_all('tr')
            for row in rows:
                cols = row.find_all('td')
                cols = [ele.text.strip() for ele in cols]
                #data.append([ele for ele in cols if ele]) # Get rid of empty values
                data.append([ele for ele in cols ]) 
    
    df = pd.DataFrame(data)
    df.columns = dataheader
    print(df)
    df.to_csv("aaa.csv")

    
 
    #dir(bs4)
def test():
    test_string = """<a href="javascript:ScoreDetailPopup('1','20170326','9');">9</a>"""
    print(test_string)
    pattern = "'(\d+)','(\d+)','(\d+)'"
    r = re.compile(pattern)
    match = r.search(str(test_string))
    print(match.group(0))
    print(match.group(1))
    print(match.group(2))
    print(match.group(3))

    

mover()
#test()
        
```
